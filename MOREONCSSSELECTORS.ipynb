```
<!DOCTYPE html>
<html lang="en">
 <head>
  <meta charset="UTF-8" />
  <meta name="viewport" content="width=device-width, initial-scale=1.0" />
  <title>Test</title>
   <style>
      /* A. Apply styles to the HTML and link it */
body {
    font-family: Arial, sans-serif;
}

/* B. Using Combinator Selectors */

/* Target all anchor tags and change text color to red */
nav a {
    color: red;
}

/* Target only the first <h2> tag and change text and background color */
h2:first-of-type {
    color: white;
    background-color: blue;
}

/* Select the first <p> tag and change text color to red */
p:first-of-type {
    color: red;
}

/* C. Using Attribute Selectors */

/* Target anchor tags with "href" attribute and change text color to red, remove text underline */
a[href] {
    color: red;
    text-decoration: none;
}

/* Target anchor tag with class name "active", change text color, and add underline */
a.active {
    color: blue;
    text-decoration: underline;
}

/* Target anchor tag where target attribute is "_self" and change background color */
a[target="_self"] {
    background-color: yellow;
}

/* D. Using Pseudo Selectors */

/* Target the last <p> tag and add "::after" with " - i.e. three dots" */
p:last-of-type::after {
    content: " ...";
}

/* Target the bullet points and change color without affecting the last item */
ul li:not(:last-child) {
    color: green;
}

/* Target the second <p> inside the second <p> and change text color and font size */
p:nth-of-type(2) p {
    color: red;
    font-size: 30px;
}
   </style>
</head>
<body>
  <header>
     <nav>
        <h2>TOI</h2>
        <div>
           <h2>Menu</h2>
            <ul>
               <li><a href="#" target="_blank">Home</a></li>
               <li><a href="#" class="active" target="_self">About</a></li>
               <li><a href="#" target="_blank">Services</a></li>
               <li><a href="#" target="_blank">Contact</a></li>
            </ul>
        </div>

         <p>The times of India</p>
         <p>
             Lorem ipsum dolor sit amet consectetur adipisicing elit. Consequatur, maxime a. Deserunt distinctio doloremque repudiandae excepturi fugit repellat nostrum culpa!
         </p>

        <p>read articles</p>
     </nav>
    </header>
</body>
</html>
```

```
<!DOCTYPE html>
<html lang="en">
<head>
  <meta charset="UTF-8" />
  <meta name="viewport" content="width=device-width, initial-scale=1.0" />
  <title>Styled Navigation Menu</title>
  <style>
    /* Remove bullet points and text underlining using Descendant Combinator */
nav ul {
    list-style: none; /* Removes bullet points */
    padding: 0; /* Removes default padding */
    margin: 0; /* Removes default margin */
}

nav ul li a {
    text-decoration: none; /* Removes underline from links */
    color: blue; /* Keeps the text blue as per the expected output */
}

/* Display menu list items inline using Child Combinator */
nav > ul > li {
    display: inline; /* Makes the list items appear in a row */
    margin-right: 15px; /* Adds spacing between items */
}

/* Use Adjacent Sibling Combinator to add margin between menu items */
nav ul li + li {
    margin-left: 10px;
}
  </style>
</head>
<body>
<header>

    <nav>
     <ul>

       <li><a href="#">Home</a></li>

       <li><a href="#">About</a></li>

       <li><a href="#">Services</a></li>

       <li><a href="#">Portfolio</a></li>

       <li><a href="#">Contact</a></li>

     </ul>

    </nav>

   </header>
   <main>

     <!--Your content here -->
   </main>
  </body>
</html>
```

```
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Button with Pseudo Elements</title>
    <style>
        /* Styling the button */
        .btn {
            position: relative;
            background-color: #4A90E2;
            color: white;
            font-size: 18px;
            padding: 12px 24px;
            border: none;
            border-radius: 8px;
            cursor: pointer;
            transition: 0.3s ease-in-out;
            font-weight: bold;
        }

        /* Adding pseudo-elements before and after */
        .btn::before {
            content: "Click ";
            font-size: 16px;
            font-weight: normal;
        }

        .btn::after {
            content: " Now";
            font-size: 16px;
            font-weight: normal;
        }

        /* Hover effect */
        .btn:hover {
            background-color: #357ABD;
            box-shadow: 0 4px 8px rgba(0, 0, 0, 0.2);
        }

        /* Centering the button */
        .container {
            display: flex;
            justify-content: center;
            align-items: center;
            height: 100vh;
            background-color: #f4f4f4;
        }
    </style>
</head>
<body>

    <div class="container">
        <button class="btn">PW Skills</button>
    </div>

</body>
</html>
```
